### Define functions
* mse: takes two images and computes the mean squared error between them
* compare_images: computes the MSE and structural similarity index and outputs the images with the scores

Images must have the same dimensions

````python
def mse(imageA, imageB):
    # the 'Mean Squared Error' between the two images is the
    # sum of the squared difference between the two images;
    # NOTE: the two images must have the same dimension
    err = np.sum((imageA.astype("float") - imageB.astype("float")) ** 2)
    err /= float(imageA.shape[0] * imageA.shape[1])
    
    # return the MSE, the lower the error, the more "similar"
    # the two images are
    return err

def compare_images(imageA_name, imageB_name):
    
    title = "%s vs %s" % (imageA_name, imageB_name)
    
    # if images have same url, don't need to compare
    if imageA_name.split("-")[1] == imageB_name.split("-")[1]:
        return
    else:
        print("%s\n%s\n" % (imageA_name.split("-")[1], imageB_name.split("-")[1]))
#         print("different")
    
    # get images from dictionary
    imageA = images[imageA_name]
    imageB = images[imageB_name]
    
    # if images have different dimensions, shrink the larger one
    if imageA.size > imageB.size:
        dim = (imageB.shape[1], imageB.shape[0])
        imageA = cv2.resize(imageA, dim, interpolation = cv2.INTER_AREA)
    else:
        dim = (imageA.shape[1], imageA.shape[0])
        imageB = cv2.resize(imageB, dim, interpolation = cv2.INTER_AREA)
        
    # compute the mean squared error and structural similarity index for the images
    m = mse(imageA, imageB)
    s = ssim(imageA, imageB)
    
    if s > 0.2:
        print title + "\nMSE: %.3f, SSIM: %.3f" % (m, s)

        # setup the figure
        fig = plt.figure(title)
#         plt.suptitle("MSE: %.2f, SSIM: %.2f" % (m, s))

        # show first image
        ax = fig.add_subplot(1, 2, 1)
        plt.imshow(imageA, cmap = plt.cm.gray)
        plt.axis("off")

        # show the second image
        ax = fig.add_subplot(1, 2, 2)
        plt.imshow(imageB, cmap = plt.cm.gray)
        plt.axis("off")

        # show the images
        plt.show()
````
